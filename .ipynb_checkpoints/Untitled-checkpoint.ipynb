```python
def plot_fft(fft):
    fig, axes =plt.subplots(nrows=2,ncols=5,sharex=False,sharey=True,figsize=(20,5))
    fig.suptitle('Fourier Transforms',size=16)
    i=0
    for x in range(2):
        for y in range(5):
            data = list(fft.value())[i]
            y,freq = data[0],data[1]
            axes[x,y].set_title(list(ftt.keys())[i])
            axes[x,y].plot(freq,y)
            axes[x,y].get_xaxis().set_visible(False)
            axes[x,y].get_yaxis().set_visible(False)
            i += 1
```

```python
# Path to the main data directory
data_dir = "data"

# Subdirectories for scream and non_scream files
categories = ['non_scream', 'scream']

# Dictionary to store file paths and labels
audio_data = []

# Traverse through the directories
for category in categories:
    folder_path = os.path.join(data_dir, category)
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".wav"):
            file_path = os.path.join(folder_path, file_name)
            label = category  # Assign label based on the folder name
            audio_data.append({"file_path": file_path, "label": label})

# Convert to DataFrame for easier manipulation
audio_df = pd.DataFrame(audio_data)
print(audio_df)
```

```python
# Save the DataFrame to a CSV file
csv_file_path = "audio_data.csv"
audio_df.to_csv(csv_file_path, index=False)
```

```python
# Print the DataFrame and the path of the saved CSV
print(audio_df)
print(f"CSV file saved at: {csv_file_path}")
```

```python
df = pd.read_csv('audio_data.csv')
```

```python
print(df)
```

```python
df['length'] = None  # Initialize the 'length' column with None or NaN

# Step 3: Iterate over the rows to calculate the length of each audio file
for index, row in df.iterrows():
    file_path = row['file_path']  # Get the file path from the DataFrame
    try:
        rate, signal = wavfile.read(file_path)  # Read the .wav file
        df.at[index, 'length'] = signal.shape[0] / rate  # Calculate length and update the DataFrame
    except FileNotFoundError:
        print(f"File not found: {file_path}")  # If file is not found, print a message
    except OSError as e:
        print(f"Error reading file {file_path}: {e}")  # Handle other errors during file reading
```

```python
# Step 4: Save the updated DataFrame with the 'length' column back to the CSV
csv_file_path = 'updated_audio_data.csv'
df.to_csv(csv_file_path, index=False)  # Save the DataFrame to CSV
```

```python
# Step 5: Print the updated DataFrame and the path of the saved CSV
print(df)
print(f"CSV file saved at: {csv_file_path}")
```

```python
# Create the classes list
classes = list(np.unique(df.label))
print("Unique classes:", classes)

# Calculate the average length per class
class_dist = df.groupby(['label'])['length'].mean()

# Display the result
print("Average length per class:")
print(class_dist)
```

```python
import matplotlib.pyplot as plt

# Assuming class_dist is already defined
fig, ax = plt.subplots()  # Correct way to get both figure and axis
ax.set_title('Class Distribution', y=1.08)
ax.pie(class_dist, labels=class_dist.index, autopct='%1.1f%%', 
       shadow=False, startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
def calc_fft(y,rate):
    n=len(y)
    freq = np.fft.rfftfreq(n,d=1/rate)
    y = abs(np.fft.rfft(y)/n)
    return(y,freq)
```

```python
import matplotlib.pyplot as plt

def plot_signal(signals):
    num_signals = len(signals)
    num_cols = 5  # Fixed number of columns
    num_rows = (num_signals // num_cols) + (1 if num_signals % num_cols != 0 else 0)  # Calculate rows dynamically
    
    # Create subplots with dynamic rows
    fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, sharex=False, sharey=True, figsize=(20, 5))
    fig.suptitle('Time Series', size=16)
    
    # Flatten axes for easier iteration (if more than one row)
    axes = axes.flatten()
    
    for i, (signal_name, signal_data) in enumerate(signals.items()):
        axes[i].set_title(signal_name)
        axes[i].plot(signal_data)
        axes[i].get_xaxis().set_visible(False)
        axes[i].get_yaxis().set_visible(False)
    
    # Hide any extra subplots if there are fewer signals than subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()
# Assuming the signals dictionary is already populated
plot_signal(signals)

def plot_fbank(fbank):
    num_signals = len(fbank)
    num_cols = 5  # Fixed number of columns
    num_rows = (num_signals // num_cols) + (1 if num_signals % num_cols != 0 else 0)  # Calculate rows dynamically
    
    # Create subplots with dynamic rows
    fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, sharex=False, sharey=True, figsize=(20, 5))
    fig.suptitle('Log Mel-Filter Bank', size=16)
    
    # Flatten axes for easier iteration (if more than one row)
    axes = axes.flatten()
    
    for i, (signal_name, bank_data) in enumerate(fbank.items()):
        axes[i].set_title(signal_name)
        axes[i].imshow(bank_data, cmap='hot', interpolation='nearest')
        axes[i].get_xaxis().set_visible(False)
        axes[i].get_yaxis().set_visible(False)
    
    # Hide any extra subplots if there are fewer signals than subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()

# Assuming the fbank dictionary is already populated
plot_fbank(fbank)

import matplotlib.pyplot as plt

def plot_fft(fft):
    num_signals = len(fft)
    num_cols = 5  # Fixed number of columns
    num_rows = (num_signals // num_cols) + (1 if num_signals % num_cols != 0 else 0)  # Calculate rows dynamically
    
    # Create subplots with dynamic rows
    fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, sharex=False, sharey=True, figsize=(20, 5))
    fig.suptitle('FFT', size=16)
    
    # Flatten axes for easier iteration (if more than one row)
    axes = axes.flatten()
    
    for i, (signal_name, fft_data) in enumerate(fft.items()):
        axes[i].set_title(signal_name)
        y, freq = fft_data[0], fft_data[1]
        axes[i].plot(freq, y)
        axes[i].get_xaxis().set_visible(False)
        axes[i].get_yaxis().set_visible(False)
    
    # Hide any extra subplots if there are fewer signals than subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()

# Assuming the fft dictionary is already populated
plot_fft(fft)

import matplotlib.pyplot as plt

def plot_mfccs(mfccs):
    num_signals = len(mfccs)
    num_cols = 5  # Fixed number of columns
    num_rows = (num_signals // num_cols) + (1 if num_signals % num_cols != 0 else 0)  # Calculate rows dynamically
    
    # Create subplots with dynamic rows
    fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, sharex=False, sharey=True, figsize=(20, 5))
    fig.suptitle('MFCC', size=16)
    
    # Flatten axes for easier iteration (if more than one row)
    axes = axes.flatten()
    
    for i, (signal_name, mfcc_data) in enumerate(mfccs.items()):
        axes[i].set_title(signal_name)
        axes[i].imshow(mfcc_data, cmap='hot', interpolation='nearest')
        axes[i].get_xaxis().set_visible(False)
        axes[i].get_yaxis().set_visible(False)
    
    # Hide any extra subplots if there are fewer signals than subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()

# Assuming the mfccs dictionary is already populated
plot_mfccs(mfccs)
```

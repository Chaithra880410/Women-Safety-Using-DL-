```python
# Path to the main data directory
data_dir = "data"

# Subdirectories for scream and non_scream files
categories = ['non_scream', 'scream']

# Dictionary to store file paths and labels
audio_data = []

# Traverse through the directories
for category in categories:
    folder_path = os.path.join(data_dir, category)
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".wav"):
            file_path = os.path.join(folder_path, file_name)
            label = category  # Assign label based on the folder name
            audio_data.append({"file_path": file_path, "label": label})

# Convert to DataFrame for easier manipulation
audio_df = pd.DataFrame(audio_data)
print(audio_df)
```

```python
# Save the DataFrame to a CSV file
csv_file_path = "audio_data.csv"
audio_df.to_csv(csv_file_path, index=False)
```

```python
df['length'] = None  # Initialize the 'length' column with None or NaN

# Step 3: Iterate over the rows to calculate the length of each audio file
for index, row in df.iterrows():
    file_path = row['file_path']  # Get the file path from the DataFrame
    try:
        rate, signal = wavfile.read(file_path)  # Read the .wav file
        df.at[index, 'length'] = signal.shape[0] / rate  # Calculate length and update the DataFrame
    except FileNotFoundError:
        print(f"File not found: {file_path}")  # If file is not found, print a message
    except OSError as e:
        print(f"Error reading file {file_path}: {e}")  # Handle other errors during file reading
```

```python
# Step 4: Save the updated DataFrame with the 'length' column back to the CSV
csv_file_path = 'updated_audio_data.csv'
df.to_csv(csv_file_path, index=False)  # Save the DataFrame to CSV
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Envelope function to create a mask based on the signal's envelope
def envelope(signal, rate, threshold):
    mask = []
    
    # Convert signal to pandas Series for easy manipulation
    signal = pd.Series(signal).apply(np.abs)
    
    # Calculate rolling mean with window size based on rate
    y_mean = signal.rolling(window=int(rate / 10), min_periods=1, center=True).mean()
    
    # Create a mask based on threshold comparison with the rolling mean
    for mean in y_mean:
        if mean > threshold:
            mask.append(True)
        else:
            mask.append(False)
    
    return mask

# Function to plot signals (both original and filtered)
def plot_signal_comparison(original_signal, filtered_signal, signal_type):
    plt.figure(figsize=(12, 6))

    # Plot original signal
    plt.subplot(2, 1, 1)
    plt.plot(original_signal, label='Original Signal')
    plt.title(f'Original {signal_type}')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.legend()

    # Plot filtered signal
    plt.subplot(2, 1, 2)
    plt.plot(filtered_signal, label='Filtered Signal', color='r')
    plt.title(f'Filtered {signal_type} (Envelope Applied)')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Function to apply the envelope detection to all signal types (time series, Mel-spectrogram, FFT)
def apply_envelope_to_signals(signals, rate, threshold=0.0005):
    filtered_signals = {}

    for signal_type, signal_data in signals.items():
        # Apply envelope to each signal
        print(f"Applying envelope detection to {signal_type}...")
        mask = envelope(signal_data, rate, threshold)
        
        # Filter the signal using the mask
        filtered_signal = np.array(signal_data)[mask]
        
        # Plot comparison of original and filtered signal
        plot_signal_comparison(signal_data, filtered_signal, signal_type)
        
        # Store the filtered signal for future use (e.g., CNN input)
        filtered_signals[signal_type] = filtered_signal

    return filtered_signals

# Example signals (replace with your actual signal data)
time_series = np.random.randn(1000)  # Replace with actual time series data
mel_spectrogram = np.random.randn(1000)  # Replace with actual Mel-spectrogram data
fft_signal = np.random.randn(1000)  # Replace with actual FFT data

# Dictionary of signals (each signal type will be processed)
signals = {
    'time_series': time_series,
    'mel_spectrogram': mel_spectrogram,
    'fft': fft_signal
}

# Apply envelope to all signals
rate = 1000  # Sampling rate (replace with actual rate)
filtered_signals = apply_envelope_to_signals(signals, rate)

# The filtered_signals dictionary contains the masked signals ready for CNN input
print(filtered_signals)
```
